```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage.filters import gaussian_filter

image = imageio.imread('download.jpg')
image_blurred = np.copy(image).astype(np.float64)
image_blurred /= 255
image_blurred = gaussian_filter(image_blurred, sigma=3)
image_blurred *= 255
image_blurred = np.round(image_blurred).astype(np.uint8)

plt.close('all')
fig, ax = plt.subplots(2,2, figsize=(10,6))

ax[0][0].imshow(image, cmap='gray')
ax[0][0].set_title('Original')
ax[1][0].hist(image.ravel(), 255, density=True)

ax[0][1].imshow(image_blurred, cmap='gray')
ax[0][1].set_title('Gaussia Filtered')
ax[1][1].hist(image_blurred.ravel(), 256, density=True)
plt.show()
```
